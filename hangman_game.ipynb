```python
#import libraries
import random
from collections import Counter


#add words in the variable
some_word_str="apple,banana,strawberry,mani,cuckoo,chintu,gabbu,mantu,chikki"

#split the words by comma
word_list=some_word_str.split(',')

#choose random words form the word list
word=random.choice(word_list)

if __name__=="__main__":
  print('guess the word!')

  for i in word:
    print('_',end=' ')
  print()


  playing=True

  letter_guessed=''
  chances=len(word)+2

  correct=0
  flag=0

  try:

    while(chances!=0) and flag==0:
      print()
      chances-=1

      try:
        guess=str(input('enter a letter to guess: '))

      except:
        print('enter only letter')
        continue

      #if the letter guessed is not alphabet (1,@,5 etc)

      if not guess.isalpha():
        print('enter only letter')
        continue

      #if the input len is greater than 1
      elif len(guess)>1:
        print('enter only a single letter')
        continue
     #Has the user already guessed this letter before?
      elif guess in letter_guessed:
        print('you have already guessed that letter')
        continue


      if guess in word:

        k=word.count(guess)
        for _ in range(k):
          letter_guessed+=guess

      # This block was causing the IndentationError
      # It should be at the same indentation level as the 'if guess in word:' statement
      print('Current word status:')
      for char in word:
        if char in letter_guessed and (Counter(letter_guessed)!=Counter(word)):

          print(char,end=' ')
          correct+=1

        elif (Counter(letter_guessed)==Counter(word)):

          print("the word is:",end=' ')
          print(word)
          flag=1
          print("congractulation")
          break # Exit the inner for loop
        else:
          print('_',end=' ')
      print()

    if chances<=0 and (Counter(letter_guessed)!=Counter(word)):
      print()
      print("you lost")
      print(f"The word was: {word}") # Using an f-string for clarity
  except KeyboardInterrupt: # Corrected typo
    print()
    print('bye')
    exit()
```
