```python
import random
print("hi welcome to the game\nlets start")
low= int(input("enter lower bound:"))
high=int(input("enter upper bound:"))

print(f"now you have 7 the chances to guess the no. between {low} and {high}")
num=random.randint(low,high)
ch=7
gu=0

while gu<ch:
  gu+=1
  guess=int(input("enter the no."))
  if guess==num:
    print(f"your guess no.{guess}is correct ")
    break

  elif gu>ch and guess!=num:
    print("better luck next time")

  elif guess>num:
    print("guess is to high")

  else:
    print("guess is too low")
```

```python
import random

word=["ritika","cuckoo","mani","chintu","mintu","chikki","gabbu"]

name = input("welcom to the game enter your name:")
print(f"hi {name} lets start")

gu=random.choice(word)


turns=12
guess=''

while turns>0:
  failed=0

  for char in gu:
    if char in guess:
      print(char, end=" ")

    else:
      print("_")
      failed+=1

  if failed==0:
    print("win")
    print(f"the word is {gu}")
    break;


  print()
  g=input("guess the char")
  guess+=g


  if g not in gu:

        turns -= 1
        print("Wrong")
        print("You have", + turns, 'more guesses')

        if turns == 0:
            print("You Loose")
```
