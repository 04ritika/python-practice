```python
class OddEven:
  l=[1,2,3,4,5,6,7,8,9]

  def ext_even(self,l):
    l1=[]
    for i in l:
      if i%2==0:
        l1.append(i)
    return l1

  def ext_odd(self,l):
    l1=[]
    for i in l:
      if i%2!=0:
        l1.append(i)

    return l1


o1=OddEven()
o1.l
o1.ext_even(o1.l)
o1.ext_odd(o1.l)
```

```python
#private
class student:
  def __init__(self,name,degree):
    self.name=name    #public variable
    self.__degree=degree  #private variable

  def show(self):
    print(self.name,self.__degree)

  #private method
  def __private_method(self):
    print("this is a private method")

  def access_private_method(self):
    self.__private_method()

stud=student("ram","master")
#stud.private_method()   #gives error


#acess private method using another method name
stud.access_private_method()
```

```python
#__eq__
class point:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def __eq__(self,other):
    return self.x==other.x and self.y==other.y


p1=point(1,2)
p2=point(1,3)
print(p1==p2)


#__add__

class point:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def __add__(self,other):
    # This should return a new point object, not an integer
    return point(self.x + other.x, self.y + other.y)


p1=point(1,2)
p2=point(1,3)
p3=(p1+p2)
print(p3.x,p3.y)
```
