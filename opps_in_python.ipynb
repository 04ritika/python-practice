```python
class OddEven:
  l=[1,2,3,4,5,6,7,8,9]

  def ext_even(self,l):
    l1=[]
    for i in l:
      if i%2==0:
        l1.append(i)
    return l1

  def ext_odd(self,l):
    l1=[]
    for i in l:
      if i%2!=0:
        l1.append(i)

    return l1


o1=OddEven()
o1.l
o1.ext_even(o1.l)
o1.ext_odd(o1.l)
```
